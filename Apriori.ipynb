```python
Historic = pd.read_csv('/home/madhav/Documents/Cargo_Virgin_Atlantis/apriori.csv',header= None)
```

```python
Historic = Historic.dropna()
Historic.index = np.arange(len(Historic))
```

```python
Historic.shape
```

```python
Historic_100 = Historic.head(100)
```

```python
records = []  
for i in range(0, 100):  
    records.append([str(Historic_100.values[i,j]) for j in range(1, 5)])
```

# Applying Apriori

```python
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)  
association_results = list(association_rules)
```
